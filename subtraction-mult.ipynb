```python
model_name = 'generalized_mult'
EPOCHS = 500
BATCH_SIZE = 4
num_digits = 3
range_min = 0
range_max = 999
num_samples = 400
```

```python
def sep(x,num_digits):
    x = np.array([int(i) for i in str(abs(x))])
    while (x.shape[0] < num_digits):
        x = np.insert(x,0,0)
    return x
```

```python
x_train_a = np.random.randint(low = range_min,high = range_max,size =num_samples)
x_train_b = np.random.randint(low = range_min,high = range_max,size =num_samples)

x_train = np.empty((num_samples,2,num_digits),dtype = np.int32)

y_train_a = x_train_a - x_train_b
y_train = np.empty((num_samples,num_digits),dtype = np.int32)

for i in range(num_samples):
    x_train[i,0] = sep(x_train_a[i],num_digits)
    
    x_train[i,1] = sep(x_train_b[i],num_digits)
    y_train[i] = sep(y_train_a[i],num_digits)
    
print('\t',x_train.shape)
print('\t',y_train.shape)
```

```python
def net(inputs):
    input_data = Input(shape = inputs)
    out = Flatten()(input_data)
#     out = Dense(6,activation='tanh')(out)
#     out = Dense(6,activation='tanh')(out)
#     out = Dense(3,activation='tanh')(out)
#     out = Dense(3)(out)
#     out = NALU(6)(out)
#     out = NALU(9)(out)
    out = NALU(3)(out)
    
    return Model(inputs = input_data,outputs = out)
```

```python
model = net((2,3))
model.compile(loss = 'MSE',optimizer = 'Adam')
model.summary()
```

```python
history =  model.fit(x_train,y_train,validation_split=0.2, epochs = EPOCHS, batch_size= BATCH_SIZE, verbose = 1)
```
