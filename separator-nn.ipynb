```python
import sys
sys.path.insert(0, 'C:/Users/Ammar/Documents/CV LAB internship/Sproj/NALU_files')
from nalu import NALU
from nac import NAC

# import Keras items
from keras.models import Model
from keras.layers import Dense,Lambda,Concatenate,Input,Activation,Layer,Dropout
# from keras import backend as K
from keras.utils import to_categorical,plot_model

# import other libs
import numpy as np
# from itertools import combinations_with_replacement,product
from sklearn.model_selection import train_test_split
```

```python
model_name = 'separator'
EPOCHS = 1000
BATCH_SIZE = 4
range_min = 100
range_max = 999
```

```python
def sep(x):
    return np.array([int(i) for i in str(x)])
```

```python
x = np.arange(range_min,range_max)
x_train,x_test = train_test_split(x)
y_train = np.array([sep(i) for i in x_train])
y_test = np.array([sep(i) for i in x_test])
```

```python
def net(inputs):
    input_data = Input(shape = inputs)
    out = Dense(20,activation = 'relu')(input_data)
#     out = NALU(20)(input_data)
#     out = NAC(10)(out)
#     out = NALU(10)(out)
#     out = NALU(10)(out)
    out = NAC(3)(out)
#     out = Dense(82,activation='softmax')(out)

    return Model(inputs = input_data,outputs = out)
```

```python
model = net((1,))
model.compile(loss = 'logcosh',optimizer = 'Adam',metrics = ['accuracy'])
model.summary()
```

```python
fitted = model.fit(x_train, y_train, validation_data = (x_test,y_test),
          batch_size=BATCH_SIZE, epochs=5,
          verbose=1)
```

```python
out = model.predict(x_test[:100])
for i in range(100):
    print ('\t',x_test[i],' = ',y_test[i,:] )
    print('\t predicted: ', out[i])
```
